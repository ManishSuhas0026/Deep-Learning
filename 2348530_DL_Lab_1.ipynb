```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
class Perceptron:
    def __init__(self, input_size):
        self.weights = np.zeros(input_size + 1)

    def predict(self, inputs):
        summation = np.dot(inputs, self.weights[1:]) + self.weights[0]
        return 1 if summation > 0 else 0

    def train(self, training_inputs, labels, learning_rate=0.1, epochs=100):
        errors = []
        for _ in range(epochs):
            total_error = 0
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                error = label - prediction
                total_error += np.abs(error)
                self.weights[1:] += learning_rate * error * inputs
                self.weights[0] += learning_rate * error
            errors.append(total_error)
        return errors


def plot_errors(errors):
    plt.plot(errors)
    plt.xlabel('Epochs')
    plt.ylabel('Total Errors')
    plt.title('Total Errors vs Epochs')
    plt.show()
```

```python
training_inputs = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
labels_and = np.array([0, 0, 0, 1])
labels_or = np.array([0, 1, 1, 1])
labels_nand = np.array([1, 1, 1, 0])
labels_xor = np.array([0, 1, 1, 0])
```

```python
perceptron_and = Perceptron(2)
errors_and = perceptron_and.train(training_inputs, labels_and)
print("AND Function Weights:", perceptron_and.weights)
plot_errors(errors_and)

perceptron_or = Perceptron(2)
errors_or = perceptron_or.train(training_inputs, labels_or)
print("OR Function Weights:", perceptron_or.weights)
plot_errors(errors_or)

perceptron_nand = Perceptron(2)
errors_nand = perceptron_nand.train(training_inputs, labels_nand)
print("NAND Function Weights:", perceptron_nand.weights)
plot_errors(errors_nand)

perceptron_xor = Perceptron(2)
errors_xor = perceptron_xor.train(training_inputs, labels_xor)
print("XOR Function Weights:", perceptron_xor.weights)
plot_errors(errors_xor)
```

```python
def f(x):
    return x**2 - 2*x + 2

def f2(x, y):
    return (1 - x)**2 + 100*(y - x**2)**2
```

```python
def gradient_descent(f, initial_x, learning_rate=0.1, threshold=1e-5):
    x = initial_x
    iterations = 0
    while True:
        gradient = 2 * x - 2
        new_x = x - learning_rate * gradient
        if abs(new_x - x) < threshold:
            break
        x = new_x
        iterations += 1
    return x, iterations
```

```python
# Minimization for f(x) = x^2 - 2x + 2
min_x, iterations = gradient_descent(f, initial_x=0)
print("Minimum for f(x) = x^2 - 2x + 2:", min_x)
print("Iterations:", iterations)
```
